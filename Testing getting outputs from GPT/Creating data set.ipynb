```python
import pandas as pd
```

```python
#Create an example dataframe
df = pd.DataFrame()
df['Question'] = ["Question"]
df['Output'] = ["Output"]
df['FullOutput'] = ["FullOutput"]
df['Date'] = ["Date o/ output"]
df['Score'] = [0]


print(df)
```

```python
#Append a row to the dataframe
def append_row(Question, Output, FullOutput, Date, Score):
    global df
    df = pd.concat([df, pd.DataFrame.from_records([{'Question' : Question,  'Output' : Output, 'FullOutput' : FullOutput, 
                                                    'Date' : Date, 'Score' : Score}])], ignore_index = True)
```
